````python
import pandas as pd
import matplotlib.pyplot as plt
````

# Gather data

## Pure penalty

```
moose -i frictionless.i Problem/type=FEProblem Outputs/file_base=penalty
```

## Augmented Lagrange

```
moose -i frictionless.i Outputs/file_base=al UserObjects/weighted_gap_uo/augmented_lagrange_predictor_scale=0
```

### Linear prediction

```
moose -i frictionless.i  Outputs/file_base=al0.5 UserObjects/weighted_gap_uo/augmented_lagrange_predictor_scale=0.5
moose -i frictionless.i  Outputs/file_base=al0.75 UserObjects/weighted_gap_uo/augmented_lagrange_predictor_scale=0.75
moose -i frictionless.i  Outputs/file_base=al1.0 UserObjects/weighted_gap_uo/augmented_lagrange_predictor_scale=1     
```

````python
penalty = pd.read_csv('penalty.csv')
al = pd.read_csv('al.csv')
al05 = pd.read_csv('al0.5.csv')
al1 = pd.read_csv('al1.0.csv')
al075 = pd.read_csv('al0.75.csv')
````

````python
plt.plot(penalty['time'], -penalty['gap'])
plt.plot(al['time'], -al['gap'])
plt.plot(al05['time'], -al05['gap'])
plt.plot(al1['time'], abs(al1['gap']))
plt.plot(al075['time'], -al075['gap'])
plt.yscale('log')
````

````python
plt.plot(penalty['time'], penalty['cumulative'])
plt.plot(al['time'], al['cumulative'])
plt.plot(al05['time'], al05['cumulative'])
plt.plot(al1['time'], al1['cumulative'])
plt.plot(al075['time'], al075['cumulative'])
````

````python
plt.plot(penalty['time'], penalty['num_nl'])
plt.plot(al['time'], al['num_nl'])
plt.plot(al05['time'], al05['num_nl'])
plt.plot(al1['time'], al1['num_nl'])
plt.plot(al075['time'], al075['num_nl'])
````
