```python
%config InlineBackend.figure_formats = ['png']
%matplotlib inline
import os
import sys
import math
import shutil
import tempfile
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams.update({'font.size': 20}) # defaults to 10.0
```

```python
df1 = pd.read_csv("dynamics_creep_1_out.csv")
df2 = pd.read_csv("dynamics_creep_2_transient_out.csv");
df = pd.concat([df1,df2])
```

```python
fig = plt.figure(figsize=(12,9))
ax = fig.gca()
ax.plot(df['time'], df['T_top'], linewidth=1.0, label='top (prescribed)')
ax.plot(df['time'], df['T_center'], linewidth=1.0, label='center')

plt.title(r'Stage 1: Transient creep')
ax.legend(loc='lower right', bbox_to_anchor=(0.05,0.9))
ax.set_xlabel(r'time')
ax.set_ylabel('Temperature')
# plt.savefig('dynamics_creep_1_out.png', dpi=300, bbox_inches='tight')
# plt.savefig('dynamics_creep_1_out.pdf', bbox_inches='tight')
plt.show()
```

```python
fig = plt.figure(figsize=(12,9))
ax = fig.gca()
ax.plot(df['time'], df['delta_l'], linewidth=1.0, label='cantilever length change')
ax.plot(df['time'], df['ycreep'], linewidth=1.0, label='creep strain yy')

plt.title(r'Stage 1: Transient creep')
ax.legend(loc='upper left', bbox_to_anchor=(0.05,0.9))
# ax.set_xlabel(r'$c_{Mo}$')
# ax.set_ylabel('free energy eV/atom')
# plt.savefig('dynamics_creep_1_out.png', dpi=300, bbox_inches='tight')
# plt.savefig('dynamics_creep_1_out.pdf', bbox_inches='tight')
plt.show()
```
